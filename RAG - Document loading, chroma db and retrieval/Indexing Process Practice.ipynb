# Chunking

```python
import nltk
import re
import string
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
lemmatize = WordNetLemmatizer()
stop_words = list(set(stopwords.words('english')))
```

```python
def clean(doc):
    docs = re.sub('[^a-zA-Z]', ' ', doc)
    docs = ''.join(doc for doc in docs if doc not in string.punctuation)
    docs = nltk.word_tokenize(docs)
    docs = [doc for doc in docs if doc not in stop_words]
    docs = [lemmatize.lemmatize(doc) for doc in docs]
    return ' '.join(docs)
```

```python
document = clean(documents[0].page_content)
document
```

### Text splitting by using Recursive method

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
text_splitter = RecursiveCharacterTextSplitter(separators = ['\n\n', '\n', ' ',''],
                                               chunk_size = 200,
                                               chunk_overlap = 50)
chunks = text_splitter.create_documents([document])
```

```python
print("Type of variable:", type(chunks))
print()
print("Type of each object inside the list:", type(chunks[0]))
print()
print("Total number of documents inside chunks:", len(chunks))
print()
print("Content of chunks:", chunks)
```

```python
texts = [doc.page_content for doc in chunks]  # Assuming each chunk has a `text` attribute
```
